**(1) Create two arrays with 10000 samples generated as [Binomial](https://en.wikipedia.org/wiki/Binomial_distribution) random variables via [Numpy](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.binomial.html) similar to what we have done with Poisson random variables in the tutorial considering the following paramenters of the Binomial function:**

**(a) $n=100$ and $p=0.3$;**

**(b) $n=10000$, and $p=0.6$**


**Make the histogram (empirical) plots related to each vector and compare them with the analytical results with the probability distribution funtion given by:**

**$$\mathrm{Prob}\left[X = k \right] = \binom{n}{k}p^k(1-p)^{n-k},$$ where $k$ is the number of *successes* in $n$ trials.**

**Remember to use this [scipy.stats.binom](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.binom.html) in a similar way we did in the tutorial with Poisson.**

```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.stats import binom
from scipy.stats import poisson
style.use('bmh')
#Not to show warning messages (to keep the notebook clean)
import warnings
warnings.filterwarnings('ignore')
```

```python
# (a) n=100, p=0.3  # number of trials, probability of each trial
samples_a = np.random.binomial(n=100, p=0.3, size=10000)

# (b) n=10000, p=0.6  # number of trials, probability of each trial
samples_b = np.random.binomial(n=10000, p=0.6, size=10000)
```

```python
### Drawing a histogram (empirical distribution)
## https://matplotlib.org/3.1.1/gallery/statistics/hist.html#
# (a) n=100, p=0.3  # number of trials, probability of each trial
samples_a = np.random.binomial(n=100, p=0.3, size=10000)
n_binsa = np.max(samples_a) - np.min(samples_a)
##
plt.figure(figsize=(14,5))
plt.hist(samples_a,n_binsa,density=True)
plt.xlabel('$n$')
plt.ylabel('Frequency of $N = n$')
plt.title('Binomial Distribution (n=100, p=0.3)')
plt.show() 
```

```python
### Drawing a histogram (empirical distribution)
## https://matplotlib.org/3.1.1/gallery/statistics/hist.html#
# (b) n=10000, p=0.6  # number of trials, probability of each trial
samples_b = np.random.binomial(n=10000, p=0.6, size=10000)
n_binsb = np.max(samples_b) - np.min(samples_b)
##
plt.figure(figsize=(14,5))
plt.hist(samples_b,n_binsb,density=True)
plt.xlabel('$n$')
plt.ylabel('Frequency of $N = n$')
plt.title('Binomial Distribution (n=10000, p=0.6)')
plt.show() 
```

```python
# Plotting for (a) n=100, p=0.3
plt.figure(figsize=(12, 6))

# Empirical Histogram
plt.hist(samples_a, bins=50, density=True, alpha=0.5, color='blue', label='Empirical')

# Analytical PDF
k_values = np.arange(0, 101)
plt.plot(k_values, binom.pmf(k_values, 100, 0.3), 'o', color='red', label='Analytical')

plt.title('Binomial Distribution (n=100, p=0.3)')
plt.xlabel('k (number of successes)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting for (b) n=10000, p=0.6
plt.figure(figsize=(12, 6))

# Empirical Histogram
plt.hist(samples_b, bins=50, density=True, alpha=0.5, color='blue', label='Empirical')

# Analytical PDF - Note: Due to the large number of possible k values, we'll only plot a subset around the mean for clarity
mean = 10000 * 0.6
k_values_subset = np.arange(int(mean - 100), int(mean + 100))
plt.plot(k_values_subset, binom.pmf(k_values_subset, 10000, 0.6), 'o', color='red', label='Analytical')

plt.title('Binomial Distribution (n=10000, p=0.6)')
plt.xlabel('k (number of successes)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```
