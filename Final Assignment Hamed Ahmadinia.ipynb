# Step 7: Load and Preprocess Data, Calculate Hourly Averages, and Perform Data Analysis

```python
# Load and preprocess data
temperature_df = pd.read_csv("Vantaa Helsinki-Vantaan lentoasema Temprature 1.1.2018-1.1.2023.csv")
wind_speed_df = pd.read_csv("Vantaa Helsinki-Vantaan lentoasema Wind speed 1.1.2018-1.1.2023.csv")
humidity_df = pd.read_csv("Vantaa Helsinki-Vantaan lentoasema Relative humidity 1.1.2018-1.1.2023.csv")
consumption_df = pd.read_csv("Electricity consumption in Finland 1.1.2018 to 1.1.2023.csv")
production_df = pd.read_csv("Electricity production in Finland 1.1.2018-1.1.2023.csv")

# Convert date and time columns to datetime objects
temperature_df['DateTime'] = pd.to_datetime(temperature_df[['Year', 'm', 'd', 'Time']].astype(str).agg('-'.join, axis=1))
wind_speed_df['DateTime'] = pd.to_datetime(wind_speed_df[['Year', 'm', 'd', 'Time']].astype(str).agg('-'.join, axis=1))
humidity_df['DateTime'] = pd.to_datetime(humidity_df[['Year', 'm', 'd', 'Time']].astype(str).agg('-'.join, axis=1))
consumption_df['DateTime'] = pd.to_datetime(consumption_df['Start time UTC+03:00'])
production_df['DateTime'] = pd.to_datetime(production_df['Start time UTC+03:00'])

# Merge consumption and production data
energy_df = consumption_df[['DateTime', 'Electricity consumption in Finland']].merge(
    production_df[['DateTime', 'Electricity production in Finland']], on='DateTime', how='inner')

# Calculate hourly averages of weather factors
temperature_hourly = temperature_df.groupby(temperature_df['DateTime'].dt.hour)['Air temperature (degC)'].mean()
wind_speed_hourly = wind_speed_df.groupby(wind_speed_df['DateTime'].dt.hour)['Wind speed (m/s)'].mean()
humidity_hourly = humidity_df.groupby(humidity_df['DateTime'].dt.hour)['Relative humidity (%)'].mean()

# Merge weather factors with energy data
energy_df['Temperature (degC)'] = temperature_hourly[energy_df['DateTime'].dt.hour].values
energy_df['Wind speed (m/s)'] = wind_speed_hourly[energy_df['DateTime'].dt.hour].values
energy_df['Relative humidity (%)'] = humidity_hourly[energy_df['DateTime'].dt.hour].values

# Data analysis
correlation_matrix = energy_df.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Correlation Matrix Interpretation
print()
print("Correlation Matrix Interpretation:")
print()
print("Correlation coefficients indicate the strength and direction of linear relationships.")
print()
print("For electricity consumption:")
print("- Positive correlation with production (0.877): When production is high, consumption tends to be high.")
print("- Weak positive correlation with temperature (0.227) and wind speed (0.193).")
print("- Negative correlation with humidity (-0.231): As humidity increases, consumption tends to decrease.")
print()
print("For electricity production:")
print("- Similar positive correlation with consumption (0.877) as observed for consumption.")
print("- Weak positive correlation with temperature (0.186) and wind speed (0.159).")
print("- Negative correlation with humidity (-0.189).")
```

```python
# Linear regression analysis
X = energy_df[['Temperature (degC)', 'Wind speed (m/s)', 'Relative humidity (%)']]
y_consumption = energy_df['Electricity consumption in Finland']
y_production = energy_df['Electricity production in Finland']

# Linear Regression Summaries Interpretation
print("\nLinear Regression Summaries Interpretation:")
print("These summaries provide insights into how weather factors influence energy consumption and production.")
```
