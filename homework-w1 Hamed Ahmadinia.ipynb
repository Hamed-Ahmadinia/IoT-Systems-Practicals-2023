```python
import pandas as pd
import matplotlib.pyplot as plt

Temperature = pd.read_csv("temp-10.9.2022-11.9.2023-Turku-airport.csv", sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])

Temperature.info()
```

```python
#plotting
plt.figure(figsize=(16,6)) 
plt.plot(Temperature,color='blue', marker='x',linestyle=':') 
plt.title("Air temperature for a year (10, sep, 2022 to 11, sep, 2023), at Turku airport station") 
plt.ylabel("Temperature in [°C]") 
plt.grid(True) 
plt.show() 
```

```python
#plotting
plt.figure(figsize=(16,6)) 
# Plotting the temperatures for January 2023
plt.plot(Temperature['2023-01-01':'2023-01-31']['Air temperature (degC)'], color='blue', marker='x', linestyle=':')

# Highlighting the temperatures that are greater than 0°C
mask = Temperature['2023-01-01':'2023-01-31']['Air temperature (degC)'] > 0
plt.scatter(Temperature['2023-01-01':'2023-01-31'][mask].index, Temperature['2023-01-01':'2023-01-31'][mask]['Air temperature (degC)'], color='black', marker='o')

plt.title("Air temperature in January 2023, at Turku airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()
```

```python
#plotting
plt.figure(figsize=(16,6)) 
# Plotting the temperatures for January 2023
plt.plot(Temperature['2023-02-06':'2023-02-12']['Air temperature (degC)'], color='blue', marker='x', linestyle=':')

# Highlighting the temperatures that are greater than 0°C
mask = Temperature['2023-02-06':'2023-02-12']['Air temperature (degC)'] > 0
plt.scatter(Temperature['2023-02-06':'2023-02-12'][mask].index, Temperature['2023-02-06':'2023-02-12'][mask]['Air temperature (degC)'], color='black', marker='o')

plt.title("Air temperature for a week from 6 to 12 February 2023, at Turku airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()
```

```python
#plotting
plt.figure(figsize=(16,6)) 
# Plotting the temperatures for January 2023
plt.plot(Temperature['2023-03-01':'2023-03-01']['Air temperature (degC)'], color='blue', marker='x', linestyle=':')

# Highlighting the temperatures that are greater than 0°C
mask = Temperature['2023-03-01':'2023-03-01']['Air temperature (degC)'] > 0
plt.scatter(Temperature['2023-03-01':'2023-03-01'][mask].index, Temperature['2023-03-01':'2023-03-01'][mask]['Air temperature (degC)'], color='black', marker='o')

plt.title("Air temperature for a day 1 March 2023, at Turku airport station")
plt.ylabel("Temperature in [°C]")
plt.grid(True)
plt.show()
```

```python
import matplotlib.dates as mdates
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(18,8))

# Data range for 2023a
range_2023a = slice('2023-6-1', '2023-6-30')
# Data range for 2023b
range_2023b = slice('2023-7-1', '2022-7-31')

# Left axis for Electricity Consumption
ax1.set_xlabel("Date")
ax1.set_ylabel("Energy in [kWh]", color='black')

# Plotting 2023a data
ax1.plot(ElectricityConsumption[range_2023a], color='red', label="Consumption  Jun 2023", linestyle='--')
# Plotting 2023b data
ax1.plot(ElectricityConsumption[range_2023b], color='green', label="Consumption July 2023", linestyle='--')

ax1.tick_params(axis='y', labelcolor='black')

# Right axis for Temperature
ax2 = ax1.twinx()
ax2.set_ylabel("Temperature in [°C]", color='blue', rotation=270, va="bottom")

# Plotting 2023a temperature data
ax2.plot(Temperature[range_2023a], color='blue', label="Temperature Jun 2023", linestyle=':')
# Plotting 2023b temperature data
ax2.plot(Temperature[range_2023b], color='cyan', label="Temperature Jully 2023", linestyle=':')

ax2.tick_params(axis='y', labelcolor='blue')

# Grid, Title, and Formatting
ax1.grid(True)
plt.title("Relationship between Electricity Consumption and Outside Air Temperature for Jun & July in 2023 for my flat in Turku")

xfmt = mdates.DateFormatter('%d-%m')
ax1.xaxis.set_major_formatter(xfmt)
ax1.xaxis.set_major_locator(mdates.MonthLocator()) 

# Legend to differentiate between the two years
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
